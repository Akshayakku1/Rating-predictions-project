```python
url = "https://www.amazon.in/Zebronics-Zeb-JUDWAA-750-Wired-Keyboard/product-reviews/B07KR5P3YD/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber=1"
reviewlist = []
def get_soup(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    return soup
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')

def get_reviews(soup):
    reviews = soup.find_all('div',{'data-hook':'review'})
    try:
        for item in reviews:
            review = {
            'products' : soup.title.text.replace('Amazon.in:Customer reviews:','').strip(),   
            'title' : item.find('a',{'data-hook': 'review-title'}).text.strip(),
            'rating': float(item.find('i',{'data-hook': 'review-star-rating'}).text.replace('out of 5 stars','').strip()),
            'body':item.find('span',{'data-hook': 'review-body'}).text.strip(),    
            }    
            reviewlist.append(review)
    except:
        pass    
```

```python
for x in range(1,30):
    soup = get_soup(f'https://www.amazon.in/Zebronics-Zeb-JUDWAA-750-Wired-Keyboard/product-reviews/B07KR5P3YD/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber=1{x}')
    print(f'Getting Page : {x}')
    get_reviews(soup)
    print(len(reviewlist))
    if not soup.find('li',{'class': 'a-disabled a-last'}):
        pass
    else:
        break
```

```python
df = pd.DataFrame(reviewlist)
df.to_excel('amazon2.zebronics.keyboard.xlsx',index=False)
print('Fine')
```
